# Tester altair

```python
import pandas as pd
data = pd.DataFrame({'a': list('CCCDDDEEE'),
                     'b': [2, 7, 4, 1, 2, 6, 8, 4, 7]})
data
```

```python
from altair import Chart
Chart(data).mark_point()
```

```python
Chart(data).mark_point().encode(
    x='a',
    y = 'b'
)
```

```python
Chart(data).mark_point().encode(
    x='a',
    y = 'average(b)'
)
```

```python
Chart(data).mark_bar().encode(
    x='a',
    y = 'b'
)
```

```python
Chart(data).mark_bar().encode(
    x='average(b)',
    y = 'a'
)
```

```python
from altair import X, Y, Axis
Chart(data).mark_bar(color='lightgreen').encode(
    X('a', axis=Axis(title='category')),
    Y('b', aggregate = 'average', axis=Axis(title='avg(b) by category'))
)


```

```python
from altair import load_dataset
df = load_dataset('seattle-weather')
print(df.head())
```

```python
from altair import Chart
Chart(df).mark_tick().encode(
    x='precipitation',
)
```

```python
Chart(df).mark_line().encode(
    X('date'),
    Y('precipitation')
)
```

```python
Chart(df).mark_line().encode(
    X('date:T', timeUnit='month'),
    Y('average(precipitation)')
)
```
